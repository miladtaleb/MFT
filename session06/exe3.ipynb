<div align="right">
<h1>نمودار ستونی</h1>
نمودارهای ستونی برای مقایسه‌ی حداقل یک مجموعه از نقاط داده مؤثرند نمودار ستونی خوشه‌ای به طور ویژه در نمایش و تجزیه‌وتحلیل چندین مجموعه از داده‌ها سودمند است
</div>

<div align="right">
<h1>نمودار میله ای</h1>
برای مقایسه‌ی مفاهیم و درصد در بین عوامل یا مجموعه‌ی داده‌ها استفاده می شوند نمودارهای میله‌ای برای نمایش و مقایسه‌ی مجموعه‌های گسترده‌ی داده‌ها یا اعداد بهترند
</div>

<div align="right">
<h1>نمودار دایره ای</h1>
نمودارهای دایره‌ای برای نشان دادن و تفکیک نمونه‌ها در یک بُعد مفیدند. این نمودار به شکل یک دایره است تا رابطه‌ی بین دسته‌های اصلی و زیر مجموعه‌ی داده‌های شما را به نمایش بگذارد. بهتر است هنگام کار با گروه‌های طبقه‌بندی شده‌ی داده‌ها، یا زمانی که می‌خواهید تفاوت‌هایی را بر اساس یک متغیر واحد در بین داده‌ها نشان دهید، از این نوع از انواع نمودار استفاده کنید.
</div>

<div align="right">
<h1>نمودار دونات یا حلقه ای</h1>
این نوع از انواع نمودار برای تبیین رابطه بین نسبت گروه‌های مختلف داده مناسبند. کاربران در این حالت می‌توانند بر نسبت برش‌ها با یکدیگر تمرکز کنند. این نوع از انواع نمودار از طریق فضای خالی میانی‌شان جزئیات بیش‌تری را نسبت به نمودارهای دایره‌ای پوشش می‌دهند.
</div>

<div align="right">
<h1>نمودار خطی</h1>
با استفاده از این نمودار می توان روند مجموعه‌ای از داده‌ها را در یک دوره‌ی خاص مشاهده کرد. همچنین با این نوع از انواع نمودار می‌توانید روند چندین گروه داده‌ی مختلف را نیز با یکدیگر مقایسه کنید.
</div>

<div align="right">
<h1>نمودار مساحتی</h1>
این نمودار برای نمایش دادن روند در طی یک دوره برای یک یا چند دسته‌‌بندی یا تغییرات بین چندین گروه از داده‌ها ایدئال است. نمودارهای مساحتی دارای دو نوع اصلی هستند: نمودار مساحتی انباشته و نمودار مساحتی تکمیل شده. هر دو نوع می‌توانند ماهیت مجموعه داده‌های انتخاب شده شما را نشان دهند
</div>

<div align="right">
<h1>نمودار پراکندگی</h1>
این نمودار برای تحلیل چگونگی قرارگیری اهداف مختلف در اطراف موضوع اصلی و ابعاد مختلف آن‌ها، بیشترین کاربرد را دارد.
</div>

<div align="right">
<h1>نمودار عنکبوتی یا رادار</h1>
اگر مجموعه‌ی بزرگی از گروه‌های مختلف داده دارید، استفاده از نمودارهای عنکبوتی بهتر از نمودارهای ستونی به حساب می‌آید. نمودار رادار در بین انواع نمودار دوبعدی برای نمایش چندین گروه داده برای حداقل سه متغیر در محورها مفید است.
</div>

<div align="right">
<h1>نمودار مقایسه ای</h1>
به عنوان نمودارهای خوشه‌ای نیز شناخته می‌شوند، به طور معمول برای مقایسه‌ی حداقل دو شی، واحد یا گروه‌های داده استفاده می‌شوند. این نمودار می‌تواند مقایسه‌ای بصری از جزئیات کیفی و کمی بین دو یا چند چیز ارائه دهد
</div>
