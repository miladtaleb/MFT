<h1>ضریب همبستگی پیرسون</h1>
یکی از رایج‌ترین روش‌های اندازه‌گیری وابستگی بین دو متغیر کمّی، محاسبه ضریب همبستگی پیرسون است این روش میان دو متغیر تصادفی برابر با کوواریانس آنها تقسیم بر انحراف معیار آنها تعریف می‌شود زمانی که ضریب برابر صفر باشد، حاکی از آن است که بین دو متغیر رابطه خطی وجود ندارد
</div>

<div align="right">
<h1>ضریب همبستگی رتبه‌ای اسپیرمن</h1>
با توجه به این که ضریب پیرسون بر اساس میانگین و واریانس محاسبه می‌گردد، ممکن است در مقابل داده‌های دورافتاده، منحرف شده و میزان همبستگی را به درستی نشان ندهد. در چنین حالتی، ضریب همبستگی رتبه‌ای اسپیرمن مورد استفاده قرار می‌گیرد مانند ضریب پیرسون، ضریب همبستگی رتبه‌ای اسپیرمن میزان تمایل یک متغیر به پیروی کردن از مقدارهای متغیر دیگر را نشان می‌دهد
</div>

<div align="right">
<h1>ضریب هماهنگی کندال</h1>
مانند ضریب همبستگی اسپیرمن، ضریب هماهنگی کندال نیز به جای مقدار از ترتیب مقدارها برای اندازه‌گیری میزان وابستگی استفاده می‌کند به وسیله این شاخص، میتوان میزان همخوانی رتبه‌ها را ارزیابی نمود 
</div>
