## 4. Draw scatter diagram and describe correlation between these:
- carat vs price
- depth vs price

```python
df = pd.read_csv(get_file('sampled_data'))
df.head()
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(15,5),sharey='row')
ax1.scatter(df['carat'],df['price'])
ax1.set_title('carat vs price')
ax1.set_xlabel('carat')
ax1.set_ylabel('price')
ax2.scatter(df['depth'],df['price'])
ax2.set_title('depth vs price')
ax2.set_xlabel('depth')
print('carat and price correlation:',round(df['carat'].corr(df['price']),2))
print('depth and price correlation:',round(df['depth'].corr(df['price']),2))
```

### carat and price correlation:
There are positive correlation and close to one between carat and price so if the vlaue of money increases, the amount of price also increases. They have correlation.
### depth and price correlation:
There are negative correlation and close to zero between depth and price so they do not have correlation.
